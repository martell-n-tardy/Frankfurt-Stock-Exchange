1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
2. Convert the returned JSON object into a Python dictionary.
3. Calculate what the highest and lowest opening prices were for the stock in this period.
4. What was the largest change in any one day (based on High and Low price)?
5. What was the largest change between any two days (based on Closing Price)?

```python
# TASK 1: Collect data from the Franfurt Stock Exchange
url = 'https://www.quandl.com/api/v3/datasets/FSE/AFX_X.json?start_date=2017-01-01&end_date=2017-12-31&api_key=t8sGXjPhKLC1vLVUsUsn' 
r = requests.get(url)
```

```python
# TASK 2: Convert the returned JSON object into a Python dictionary
json_data = r.json()

# confirm object in type dict
print(type(json_data)) 
```

```python
# Task 3: Calculate what the highest and lowest opening prices were for the stock in this period

# find which value in the key 'column_names' contains info on 'Open' prices.
col_ = json_data['dataset']['column_names']
print(col_)
```

Column 2 in this dict contains the information on pricing needed. 

```python
# Isolate dict "data"
data_ = json_data['dataset']['data']
```

```python
# Get'Open' data (2nd item) from every list in 'data_'
def Extract_Open(data_):
    return [item[1] for item in data_]

openData = Extract_Open(data_)

# Remove None from new list
openData_NA = [i for i in openData if i]
print(openData_NA)
```

```python
# sort'OpenData' in ascending order to see highest and lowest opening prices
openData_NA.sort(reverse=True) 
print(openData_NA)
```

The highest and lowest opening prices were for the stock in this period are 53.11 and 34.0 respectively.

```python
# Task 4: The largest change in any one day (based on High and Low price)

# Get'high'[2] data from every list in 'data_'
def Extract_high(data_):
    return [item[2] for item in data_]

highData = Extract_high(data_)

# Remove None from new list
highData_NA = [i for i in highData if i]
print(highData_NA)
```

```python
# Get'low'[3] data from every list in 'data_'
def Extract_low(data_):
    return [item[3] for item in data_]

lowData = Extract_low(data_)

# Remove None from new list
lowData_NA = [i for i in lowData if i]
print(lowData_NA)
```

```python
# Find difference (high - low) of each day
import operator
diff = list(map(operator.sub, highData_NA,lowData_NA))
diff.sort(reverse=True)
print(diff)
```

The largest difference in a day was 2.81 based on 'High' and 'Low' prices.

```python
# Task 5: The largest change between any two days (based on Closing Price)

# Get'closing'[4] data from every list in 'data_'
def Extract_closing(data_):
    return [item[4] for item in data_]

closingData = Extract_closing(data_)

# Remove None from new list
closingData_NA = [i for i in closingData if i]
print(closingData_NA)
```

```python
# Calculate difference list
diff_list = []
for x, y in zip(closingData_NA[0::], closingData_NA[1::]):
    diff_list.append(y-x)
    diff_list.sort(reverse=True)
print(diff_list)
```

The largest change between any two days was 2.56 based on Closing Price.

```python
# Task 6: The average daily trading volume during this year

# Get'Traded Volume'[6] data from every list in 'data_'
def Extract_volume(data_):
    return [item[6] for item in data_]

volumeData = Extract_volume(data_)

# Remove None from new list
volumeData_NA = [i for i in volumeData if i]
print(volumeData_NA)
```

```python
# Calcuate the daily average:
len(volumeData_NA) # num of days recorded = 255
avg = sum(volumeData_NA)/255
print(round(avg,2))
```

The average daily trading volume during this year was 89,124.34.

```python
# Task 7: The median trading volume during this year
import statistics
statistics.median(volumeData_NA)
```

The median trading volume during this year was 76,286.
